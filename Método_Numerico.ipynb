```python
import numpy as np
import matplotlib.pyplot as plt

def regresion_lineal(x, y):
    # Calcula los coeficientes de la regresión lineal
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    xy = np.sum(x * y)
    x_squared = np.sum(x**2)
    m = (xy - len(x) * mean_x * mean_y) / (x_squared - len(x) * mean_x**2)
    n = mean_y - m * mean_x
    return m, n


y = np.array([500, 510, 520, 530, 540, 550, 560, 570, 580, 590])
```

```python
def calcular_r_cuadrado(x, y, m, n):
    # Calcula el valor estimado de y utilizando la regresión lineal
    y_estimado = m * x + n

    # Calcula la suma de cuadrados totales
    ss_tot = np.sum((y - np.mean(y))**2)

    # Calcula la suma de cuadrados de residuos
    ss_res = np.sum((y - y_estimado)**2)

    # Calcula el coeficiente de determinación (R²)
    r_cuadrado = 1 - (ss_res / ss_tot)

    return r_cuadrado
```

Aplicación del método

```python
temp_x = np.array([28.2, 29.5, 29.0, 30.3, 31.8, 31.0, 30.7, 32.2, 32.0, 33.1])
m_temp, n_temp = regresion_lineal(temp_x, y)

# Calcula los valores de y predichos por la regresión lineal
y_temp = m_temp * temp_x + n_temp

# Grafica los puntos y la recta de regresión
plt.scatter(temp_x, y, label='Datos')
plt.plot(temp_x, y_temp, color='red', label='Recta de Regresión')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Producción del cultivo (Toneladas)')
plt.legend()
plt.show()
```

```python
print(calcular_r_cuadrado(temp_x, y, m_temp ,n_temp))
```
