```python
import numpy as np
import matplotlib.pyplot as plt

def regresion_lineal(x, y):
    # Calcula los coeficientes de la regresión lineal
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    xy = np.sum(x * y)
    x_squared = np.sum(x**2)
    m = (xy - len(x) * mean_x * mean_y) / (x_squared - len(x) * mean_x**2)
    n = mean_y - m * mean_x
    return m, n


y = np.array([500, 510, 520, 530, 540, 550, 560, 570, 580, 590])
```

```python
fertilizante_x = np.array([28.2, 29.5, 29.0, 30.3, 31.8, 31.0, 30.7, 32.2, 32.0, 33.1])
m, n = regresion_lineal(fertilizante_x, y)

# Calcula los valores de y predichos por la regresión lineal
y_pred = m * fertilizante_x + n

# Grafica los puntos y la recta de regresión
plt.scatter(fertilizante_x, y, label='Datos')
plt.plot(fertilizante_x, y_pred, color='red', label='Recta de Regresión')
plt.xlabel('Fertilizante utilizado (Toneladas)')
plt.ylabel('Producción del cultivo (Toneladas)')
plt.legend()
plt.show()
```
